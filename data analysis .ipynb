# Insight 1

```python
ad=pd.merge(d_customers,d_region, on ='Region Code')
ad=pd.merge(ad,d_division, on ='Division')
ad=pd.merge(ad,d_caddress, on ='Address Number')
ad=pd.merge(ad,d_statedescription, on ='State')
ad.head()
```

```python

fig, axes= plt.subplots(figsize=(8,5))
total = float(len(ad))
ax=sns.countplot(x='Division Name',linewidth=15,data=ad)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.show()
plt.tight_layout()
```

##From above countplot we can say that 92.9% customers are domestic

```python
fig, axes= plt.subplots(figsize=(40,20))
total = float(len(ad))
ax=sns.countplot(x='state_name',linewidth=15,data=ad)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.tight_layout()
```

##From the plot we can see that 10.1% of the domestic customers are from California

```python
p=ad[ad['State'].str.contains('CA')]
fig, axes= plt.subplots(figsize=(20, 12))
total = float(len(p))
ax=sns.countplot(x='City',linewidth=15,data=p)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width()
    y = p.get_height()
    ax.annotate(percentage, (x, y),ha='center')
plt.tight_layout()
```

##Scaramento is the city with most number of customers in California with 22.8% customers of california in it.
